```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('Data/airline_passengers.csv', index_col='Month', parse_dates=True)
df.index.freq = 'MS'
```

```python
df.head()
```

```python
df.plot(figsize = (10, 4), grid = True)
```

```python
from statsmodels.graphics.tsaplots import month_plot, quarter_plot
```

```python
'''
since our data is already monthly, we can directly use month_plot
This is similar to a group-by operation by month.
'''

month_plot(df['Thousands of Passengers']);
```

```python
'''
As our data is monthly we first have to resample it to quarters.
'''
dfq = df['Thousands of Passengers'].resample(rule='Q').mean()
```

```python
quarter_plot(dfq);
```
