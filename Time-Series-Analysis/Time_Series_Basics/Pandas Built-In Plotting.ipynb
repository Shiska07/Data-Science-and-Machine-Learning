```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.read_csv('Data/df1.csv')
df2 = pd.read_csv('Data/df2.csv')
df3 = pd.read_csv('Data/df3.csv')
```

## Histograms

```python
df1['A'].plot.hist(figsize = (8, 5), grid = True, bins = 20, edgecolor = 'k')
```

## Bar Plots

```python
df2.plot.bar(figsize = (8, 5), grid = True)
```

```python
df2.plot.barh(figsize = (8, 5), grid = True)
```

```python
df2.plot.bar(stacked = True, figsize = (8, 5), grid = True)
```

## Line Plots

```python
df2.plot.line(figsize = (8, 5), grid = True, lw = 2, y = 'b', ls = '--')
```

```python
df2.plot.line(y = ['a', 'b', 'c', 'd'], figsize = (8, 5), grid = True)
```

## Area Plots

```python
df2.plot.area(y = ['a', 'b', 'c', 'd'], figsize = (8, 5), grid = True)
```

```python
df2.plot.area(stacked = False, alpha = 0.5, y = ['a', 'b', 'c', 'd'], figsize = (8, 5), grid = True)
```

## Scatter Plots

```python
# c = hue
df1.plot.scatter(figsize = (8, 5), grid = True, x = 'A', y = 'B', c = 'C', cmap = 'coolwarm')
```

```python
# choosing size instead of coloe as hue
df1.plot.scatter(figsize = (8, 5), grid = True, x = 'A', y = 'B', s = df1['C']*35, alpha = 0.6)
```

```python
# choosing size as 'C' and color for hue as 'D'
df1.plot.scatter(figsize = (8, 5), grid = True, x = 'A', y = 'B', c = 'D', cmap = 'coolwarm', s = df1['C']*35, alpha = 0.7)
```

## Modifying Line Plots

```python
ax = df2['c'].plot.line(figsize = (8, 5), grid = True, lw = 2, ls = '-.', c = 'red', title = 'Time-ser c')
ax.set(xlabel = 'index', ylabel = 'c')
```

```python
ax = df2.plot.line(figsize = (8, 5), grid = True, lw = 2, ls = '-.', title = 'Time-ser df2')
ax.set(xlabel = 'index', ylabel = 'values')
ax.legend()
```

```python
ax = df2.plot.line(figsize = (8, 5), grid = True, lw = 2, ls = '-.', title = 'Time-ser df2')
ax.set(xlabel = 'index', ylabel = 'values')
ax.legend(loc = 0, bbox_to_anchor = (1.0, 0.3))
```
