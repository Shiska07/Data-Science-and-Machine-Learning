### Stationary vs Non-Stationary Data
Stationary data has a constant mean and variance as time progresses. This means it does not have an increasing or a decreasing trend, cyclic amplitude(variance) or frequency(covariance).

#### ARIMA Parameters
1. AR(p): Autoregression component relates to the relationship of current data with past data.
2. I(d): Integrated component related to how many timesteps of diff() must be applied to convert non-stationary to stationary.
3. MA(q): Moveing Average component relates to the dependency between an observation and a residual error from an moving average model applied to lagged observations.

### ADF test to confirm stationarity

```python
from statsmodels.tsa.stattools import adfuller

# function for generating ADF interpretation report
def adf_test(series,title=''):
    """
    Pass in a time series and an optional title, returns an ADF report
    """
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC') # .dropna() handles differenced data
    
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)

    for key,val in result[4].items():
        out[f'critical value ({key})']=val
        
    print(out.to_string())          # .to_string() removes the line "dtype: float64"
    
    if result[1] <= 0.05:
        print('\nConclusion:')
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("Data has no unit root and is stationary")
    else:
        print('\nConclusion:')
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis")
        print("Data has a unit root and is non-stationary")
```

```python
adf_test(df1['Births'])
```

### ARIMA

```python
# trade inventory data will be used for this
df2.plot(figsize = (10, 4), grid = True)
```

```python
# Sice we are note sure if there is a seasonal component to this we will decompose
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
res = seasonal_decompose(df2['Inventories'], model='add')
res.plot();
```

We can see that there is a seasonal component to the data however, in comparison to the trens values they are very small.

```python
auto_arima(df2['Inventories'], seasonal=False, start_p=1, start_d=1, max_p=5, max_d=5, max_q = 5).summary()
```

```python
'''
Since the recommended d term is 1, we can use ADF test on the differentiated dataset to check for stationarity.
'''
from statsmodels.tsa.statespace.tools import diff
```

```python
df2['diff1'] = diff(df2['Inventories'], k_diff=1)
```

```python
# adf on original values
adf_test(df2['Inventories'])
```

```python
# adf on differentiated data
adf_test(df2['diff1'])
```

Here we can see that differentiating the data once makes it stationary. Therefore, d = 1 is adequate.

### Getting 'p' and 'q' components from acf and pacf plots

```python
plot_acf(df2['Inventories'], lags = 40);
```

```python
plot_pacf(df2['Inventories'], lags = 40);
```

Since the PACF plot has a sharp cut-off at 1, we will use an AR term of 1. For the MA term, we can check if the acf plot is decaying slowly and if most of the smaller lags are outside of the shaded region. If this seems to be the pattern, it is suggested to use an AR term rather than an MA term.

```python
# train test split
print(len(df2['Inventories']))
train = df2['Inventories'].iloc[:230]
test = df2['Inventories'].iloc[230:]
print(len(train))
print(len(test))
```

```python
start = len(train)
end = start + len(test) - 1
```

```python
# we will test two models
arima0 = ARIMA(train, order = (1, 1, 0)).fit()
arima1 = ARIMA(train, order = (1, 1, 1)).fit()
```

```python
pred0 = arima0.predict(start, end, typ = 'levels').rename('ARIMA (1, 1, 0)')
pred1 = arima1.predict(start, end, typ = 'levels').rename('ARIMA (1, 1, 1)')
```

```python
test.plot(figsize = (10, 5), legend = True)
pred0.plot(legend = True)
pred1.plot(legend = True, grid = True, title = 'ARIMA model performance for MA = 0 vs 1')
```

```python
from statsmodels.tools.eval_measures import rmse
```

```python
print(rmse(test, pred0))
print(rmse(test, pred1))
```
