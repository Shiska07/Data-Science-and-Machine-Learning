```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = 'C:/Users/shisk/Desktop/Projects/Data/iris.csv'
```

```python
df = pd.read_csv(path)
df.head()
```

```python
df.dropna(axis = 0, inplace = True)
df.reset_index(drop = True, inplace = True)
df.info()
```

```python
df.drop('Id', inplace = True, axis = 1)
```

```python
# visualiza the data
sns.pairplot(df)
```

```python
# we will try to create clsuters based on petal length and Sepal length
df_feat = df[['SepalLengthCm','PetalLengthCm']].copy()
df_feat.head()
```

```python
# standardize the data
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
```

```python
df_featsc = pd.DataFrame(sc.fit_transform(df_feat), columns = df_feat.columns)
df_featsc.head()
```

```python
from sklearn.cluster import KMeans
```

```python
kmc = KMeans(n_clusters = 3)
```

```python
kmc.fit(df_featsc)
```

```python
# to get labels
kmc.labels_
```

```python
# Get original data to compare with clustering results
df_orig = df[['SepalLengthCm','PetalLengthCm','Species']].copy()
```

```python
df_orig['Species'].replace(['Iris-setosa','Iris-versicolor','Iris-virginica'], [0, 1, 2], inplace = True)
df_orig.head()
```

```python
# plot original dataset
sns.scatterplot(x = df_orig['SepalLengthCm'], y = df_orig['PetalLengthCm'], hue = df_orig.Species, palette = "deep").set(title = "Original Dataset")
```

```python
# plot results from K-means Clustering
sns.scatterplot(x = df_featsc['SepalLengthCm'], y = df_featsc['PetalLengthCm'], hue = kmc.labels_, palette = "deep").set(title = "K-Means Clustering")
```
